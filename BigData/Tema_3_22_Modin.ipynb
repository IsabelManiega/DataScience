```python
# pip install modin
```

```python
%%time
import ray
ray.init()
import modin
import modin.pandas as mpd
modin.config.Engine.put("ray")
```

Lectura de csv utilizando Modin-Ray

```python
%%time
df_modin_ray = mpd.read_csv('yellow_tripdata.csv')
df_modin_ray.head()
```

```python
%%time
df_modin_ray.describe()
```
