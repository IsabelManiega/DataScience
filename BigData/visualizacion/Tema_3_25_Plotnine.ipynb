# PLOTNINE

Ggplot es una líbrería muy importante en "R" (lenguaje de programación).

En el caso de Python, lo que se va pedir es que se aprenda de forma autodidacta a hacer las gráficas para el dataset Iris con "plotnine".

Los colores no necesitan coincidir, pero la gráfica debería de ser la misma.(pétalo, sépalo).

```python
# Primero realizamos la instalación de la librería
# pip install plotnine
# Es necesario tener instalada la version de matplotlib 3.5.0, ya que con la última versión da problemas!!!
```

```python
import pandas as pd
from plotnine import *
```

```python
df = pd.read_csv("Iris.csv")
df
```

## Gráfico del Pétalo - con Plotnine

```python
# Create a simple scatter plot.
# Note, the parens wrapping the statement allow you to use `+` at the end of the line
# without escaping with a backslash.
(ggplot(df, aes('PetalWidthCm', 
                'PetalLengthCm')) +
  geom_point())
```

```python
(ggplot(df, aes('PetalWidthCm', 
                'PetalLengthCm', 
                color='Species')) +
  geom_point() +
  stat_smooth(method='lm')
)
```

```python
(ggplot(df, aes('PetalWidthCm', 
                'PetalLengthCm', 
                color='Species')) +
  geom_point() 
)
```

## Gráfico del Sépalo - con Plotnine

```python
(ggplot(df, aes('SepalWidthCm', 
                'SepalLengthCm', 
                color='Species')) +
  geom_point() 
)
```

### Más características del gráfico

```python
(ggplot(data=df,
        mapping=aes(x='SepalWidthCm', 
                    y='SepalLengthCm', 
                    color='Species'))
    + geom_point(alpha=0.5) 
    + xlab("SepalWidth (cm)") 
    + ylab("SepalLength (cm)") 
    + theme_bw()
)
```

```python
(ggplot(data=df,
        mapping=aes(x='SepalWidthCm', 
                    y='SepalLengthCm', 
                    color='Species', 
                    shape="Species"))
    + geom_point(alpha=0.5) 
    + xlab("SepalWidth (cm)") 
    + ylab("SepalLength (cm)") 
    + theme_bw()
)
```

```python
(ggplot(df, aes('PetalWidthCm', 'PetalLengthCm', color='Species')) +
  geom_point() 
)
```

### boxplot

```python
(ggplot(data=df,
        mapping=aes(x='SepalWidthCm',
                    y='SepalLengthCm',
                    color="Species"))
    + geom_boxplot()
)
```

```python
(ggplot(data=df,
        mapping=aes(x='PetalWidthCm',
                    y='PetalLengthCm',
                    color="Species"))
    + geom_boxplot()
)
```
